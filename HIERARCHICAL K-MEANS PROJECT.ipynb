```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_excel("Wholesale customers data.xlsx")
df.head()
```

```python
def circlepoints(r, n=200, noise = 20):
    return [(math.cos(2*math.pi/n*i)*r + np.random.uniform(-noise, noise), math.sin(2*math.pi/n*i)*r + np.random.uniform(-noise, noise)) for i in range(n)]
```

```python
df = pd.DataFrame(circlepoints(100, 200, 20))
df = df.append(circlepoints(300, 400, 40))
df = df.append(circlepoints(500, 700, 60))
df = df.append([(np.random.randint(-650, 650), np.random.randint(-650, 650)) for i in range(200)])
plt.scatter(df[0], df[1], s =10)
```

```python
hac = AgglomerativeClustering(n_clusters = 2)
hac.fit(df[[0,1]])
df["HAC"] = hac.labels_
plt.scatter(df[0], df[1], c = df["HAC"])
```
